```r
# Initialize empty dataframe to store results
cell_proportions_df <- data.frame()

# Define the mapping between scRNA and spatial cell types
cell_type_mapping <- list(
  'Malignant' = 'Malignant',
  'T_cells' = 'T.cells',
  'B_cells' = 'B.cells',
  'Myeloid' = 'Myeloid',
  'Other' = 'Nonmalignant.others',
  'Fibroblasts' = 'Fibroblasts',
  'Endothelial' = 'Endothelial'
)
sc_rna_key <- 0
# Loop through samples
for (x in samples){
    sc_rna_key <- sc_rna_key + 1
    print(x)
    seu <- subset(durva, orig.ident == scRNA[[sc_rna_key]])
    minFeature<-200
    maxFeature<- 9000
    minCount<- 800
    maxCount<- 40000
    maxMT<-15
    seu <- subset(seu, subset = nFeature_RNA > minFeature & nFeature_RNA < maxFeature & nCount_RNA > minCount &
                  nCount_RNA < maxCount & percent.mt < maxMT & predicted_doublets == FALSE)
    
    scRNA.AFM <- c(scRNA.AFM, seu)

    # Read and process the spatial data
    df <- read_csv(paste0("UpdatedCytoSpace/",x, "/cytospace_results/cell_type_assignments_by_spot.csv"))

    #df <- read_csv(paste0(x, "/Cytospace/cell_type_assignments_by_spot.csv"))
    
    df <- df %>% select(-`Total cells`)
    
    # Summing each column in the spatial data
    cell_sums_spatial <- colSums(df[, -1])  # Exclude the SpotID column
    total_cells_spatial <- sum(cell_sums_spatial)
    cell_proportions_spatial <- cell_sums_spatial / total_cells_spatial
    
    # Process the scRNA data
    major_groups <- seu@meta.data$Updated_CellType
    
    total_cells_scRNA <- length(major_groups)
    cell_proportions_scRNA <- table(major_groups) / total_cells_scRNA
    
    # Combine both scRNA and spatial proportions into a data frame using the mapping
    for (cell_type_scRNA in names(cell_proportions_scRNA)) {
        # Find the corresponding spatial cell type
        
        #cell_type_spatial <- cell_type_mapping[[cell_type_scRNA]]
        
        cell_type_spatial<- cell_type_scRNA
        # Match spatial data to the cell type, if present
        spatial_proportion <- ifelse(cell_type_spatial %in% names(cell_proportions_spatial), 
                                     cell_proportions_spatial[cell_type_spatial], 
                                     NA)
        
        # Create a new row and bind it to the dataframe
        new_row <- data.frame(
            x = x,
            cell_type = cell_type_scRNA,
            cell_proportions_scRNA = cell_proportions_scRNA[cell_type_scRNA],
            cell_proportions_spatial = spatial_proportion
        )
        
        # Append to the results dataframe
        cell_proportions_df <- rbind(cell_proportions_df, new_row)
    }
}

# View the final dataframe
cell_proportions_df
```

```r
# Scatter plot with cell_proportions_scRNA on x-axis and cell_proportions_spatial on y-axis
ggplot(cell_proportions_df, aes(x = cell_proportions_scRNA, y = cell_proportions_spatial, color = cell_type)) +
  geom_point(size = 3) +                      # Scatter plot points
  geom_abline(slope = 1, intercept = 0, linetype = "dashed", color = "black") +  # Line y=x
  stat_cor(method = "pearson", label.x = 0.7, label.y = 0.1) +  # Display Pearson correlation
  labs(x = "Proportion (scRNA)", y = "Proportion (Spatial)", color = "Cell Type") +  # Axis and legend labels
  theme_classic(base_size = 16) +             # Clean theme for publication
  theme(legend.position = "right") +          # Position of the legend
  scale_color_brewer(palette = "Set1") +      # Color scheme suitable for publication
  ggtitle("Correlation Between scRNA and Spatial Cell Proportions")  # Title
```

```r
# Load necessary libraries
library(ggplot2)
library(ggpubr)
library(dplyr)
options(repr.plot.width = 8, repr.plot.height = 5)

# Assuming your data is in a dataframe called df
df <- cell_proportions_df



cell_types_with_colors <- c("B.cells", "Endothelial", "Fibroblasts", "Malignant", "Myeloid", "Other", "T.cells")
colors <- c("#000000", "#efc100", "#84538e", "#f38400", "#9cbee5", "#d72828", "#d9c69c")

# Update the cell_type column: merge all unlisted cell types into "Other"
df <- df %>%
  mutate(cell_type = ifelse(cell_type %in% cell_types_with_colors, cell_type, "Other"))

# Replace underscores in B_cells and T_cells with spaces for the plot
df <- df %>%
  mutate(cell_type = recode(cell_type, 
                            'T.cells' = 'T cells', 
                            'B.cells' = 'B cells'))

# Add "Other" to the color list with the specified color
colors <- c(colors, "#d72828")
names(colors) <- c("B cells", "Endothelial", "Fibroblasts", "Malignant", "Myeloid", "Other", "T cells")

# Scatter plot with cell_proportions_scRNA on x-axis and cell_proportions_spatial on y-axis
ggplot(df, aes(x = cell_proportions_scRNA, y = cell_proportions_spatial, color = cell_type)) +
  geom_point(size = 5) +                      # Scatter plot points
  geom_abline(slope = 1, intercept = 0, linetype = "dashed", color = "black") +  # Line y=x
  labs(x = "Proportion (snRNA)", y = "Proportion (Spatial Projection)", color = "Cell Type") +  # Axis and legend labels
  theme_classic(base_size = 18) +             # Set base font size to 18 for axes and title
  theme(
    legend.position = "right",               # Position of the legend
    legend.title = element_text(size = 18),  # Set legend title font size to 18
    legend.text = element_text(size = 18),   # Set legend text font size to 18
    axis.title = element_text(size = 18),    # Set axis titles font size to 18
    axis.text = element_text(size = 18),     # Set axis labels font size to 18
    plot.title = element_text(size = 18, hjust = 0.5)  # Set title font size to 18 and center it
  ) +
  scale_color_manual(values = colors)        # Use custom colors
```

```r
# Compute MSE
mse <- mean((df$cell_proportions_scRNA - df$cell_proportions_spatial)^2)

# Fit a linear model to get R^2
model <- lm(cell_proportions_spatial ~ cell_proportions_scRNA, data = df)
r_squared <- summary(model)$r.squared

# Compute p-value for the correlation
cor_test <- cor.test(df$cell_proportions_scRNA, df$cell_proportions_spatial)
p_value <- cor_test$p.value

# Results
cat("MSE:", mse, "\n")
cat("R-squared:", r_squared, "\n")
cat("p-value:", p_value, "\n")
```
