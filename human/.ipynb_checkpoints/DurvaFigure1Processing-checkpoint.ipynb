```r
# Load necessary libraries
library(ggplot2)
library(Seurat) # Assuming Seurat is used here

# Reduce plot dimensions and increase font size for Jupyter notebook display
options(repr.plot.width = 6, repr.plot.height = 4)  # Smaller figure size

# Plotting with ggplot2
ggplot(merged_seurat@meta.data, aes(x = orig.ident, y = percent.mt, fill = orig.ident)) +
  geom_violin(color = "black") +
  geom_boxplot(width = 0.1, fill = "white", color = "black", outlier.shape = NA) +  # Adding boxplot inside violin
  theme_minimal() +
  labs(x = "orig.ident", y = "Percent Mitochondrial") +
  theme(
    text = element_text(size = 16),      # General text size
    axis.text = element_text(size = 14), # Axis tick text size
    axis.title = element_text(size = 16), # Axis labels text size
    legend.position = "none"             # Optional: hide legend if not needed
  )
scRNA.AFM <- c()
spatial.AFM <- c()
sc_rna_key <- 0
samples <- c("ZWV08", "ZWV10", "ZWV14","ZWV16")
scRNA <- c('Du13','Du27', 'Du53', 'Du55')

for (x in samples){
    sc_rna_key <- sc_rna_key + 1
    print(x)
    seu <- readRDS(list.files(path = paste0(x, "/scRNA") , 
                              pattern = "\\.rds$", full.names = TRUE)[1])
    minFeature<-200
    maxFeature<- 9000
    minCount<- 800
    maxCount<- 40000
    maxMT<-15
    seu <- subset(seu, subset = nFeature_RNA > minFeature & nFeature_RNA < maxFeature & nCount_RNA > minCount &
                nCount_RNA < maxCount & percent.mt < maxMT & predicted_doublets ==F)
  scRNA.AFM[[sc_rna_key]] <- seu
}

# Merge all the Seurat objects into one
merged_seurat <- merge(scRNA.AFM[[1]], y = scRNA.AFM[2:length(scRNA.AFM)], 
                       add.cell.ids = samples, project = "Merged_scRNA")

# Calculate the percentage of mitochondrial genes
merged_seurat[["percent.mt"]] <- PercentageFeatureSet(merged_seurat, pattern = "^MT-")

# Generate Violin plots with box plots for 'nFeature_RNA', 'nCount_RNA', and 'percent.mt'
features <- c("nFeature_RNA", "nCount_RNA", "percent.mt")
plot_list <- list()

for (feature in features) {
  p <- VlnPlot(merged_seurat, features = feature, group.by = "orig.ident", pt.size = 0) +
    geom_boxplot(width = 0.1, outlier.shape = NA) + # Adding boxplot with no outliers
    theme_minimal(base_size = 16) + # Set font size for publication
    theme(
      axis.title = element_text(size = 18), # Axis title size
      axis.text = element_text(size = 16), # Axis text size
      plot.title = element_text(size = 20), # Plot title size
      legend.text = element_text(size = 14), # Legend text size
      legend.title = element_text(size = 16) # Legend title size
    ) +
    coord_fixed(ratio = 1) + # Make plots square
    ggtitle(paste("Violin plot for", feature))
  plot_list[[feature]] <- p
  print(p)
}

# To display plots together, you can use patchwork or arrange them
combined_plot <- plot_list[["nFeature_RNA"]] + plot_list[["nCount_RNA"]] + plot_list[["percent.mt"]] +
  plot_layout(ncol = 3) & 
  theme(plot.margin = margin(5, 5, 5, 5))

# Show the combined plot
print(combined_plot)
```

```r
df <- read.csv("integrated_data/all_cell_metadata.csv")
features <- c("nFeature_RNA", "nCount_RNA", "percent.mt")

# Generate Violin plots for each feature, grouped by 'orig.ident'
plot_list <- list()

for (feature in features) {
  p <- ggplot(df, aes(x = orig.ident, y = .data[[feature]])) +
    geom_violin(trim = TRUE) +
    theme_minimal() +
    labs(title = paste("Violin plot for", feature), x = "orig.ident", y = feature) +
    theme(axis.text.x = element_text(angle = 45, hjust = 1))
  
  plot_list[[feature]] <- p
  print(p)
}

# Optionally, use patchwork to display all plots together
library(patchwork)
plot_list[["nFeature_RNA"]] + plot_list[["nCount_RNA"]] + plot_list[["percent.mt"]]
```
